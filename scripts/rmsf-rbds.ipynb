```python
# librarias
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

### Ruta
path = "/home/ahr/Desktop/spike_MD/"

### crear datasets desde archivos xvg

#orig (ax1)
x,y = np.loadtxt(path + "orig/rmsf_rbd_orig.xvg",comments=["@", "#"],unpack=True)
#delta1 (ax2)
x1,y1 = np.loadtxt(path + "delta/inicial/rmsf_rbd_inicial.xvg",comments=["@", "#"],unpack=True)
#delta3 ((ax3)
x2,y2 = np.loadtxt(path + "delta/final/rmsf_rbd_final.xvg",comments=["@", "#"],unpack=True)
#omicron 21K (ax4)
x3,y3 = np.loadtxt(path + "omicron/rmsf_rbd_omicron.xvg",comments=["@", "#"],unpack=True)
#omicron 21L (ax5)
x4,y4 = np.loadtxt(path + "omicron_21L/rmsf_rbd_omicron21L.xvg",comments=["@", "#"],unpack=True)

##crear grfica
fig, axs = plt.subplots(2, 3, figsize=(18, 8), sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0})
(ax1, ax2,ax3), (ax4, ax5,ax6) = axs

fig.suptitle('RMSF calculation in RBD of SARS-CoV-2 variants', fontsize=28) 

ax1.plot(x, y, color='grey', linewidth=3, label='Original')
ax1.set_ylabel('RMSF', fontsize=13)
ax1.legend(loc='upper left')

ax2.plot(x1, y1, 'darkturquoise', linewidth=3, label='21J Delta (1)')
ax2.legend(loc='upper left')

ax3.plot(x2, y2, 'mediumseagreen', linewidth=3, label='21J Delta (3)')
ax3.legend(loc='upper left')

ax4.plot(x3, y3, 'gold', linewidth=3, label='21K omicron')
ax4.set_ylabel('RMSF', fontsize=13)
ax4.legend(loc='upper left')

ax5.plot(x4, y4, 'goldenrod', linewidth=3, label='21L omicron')
ax5.legend(loc='upper left')

ax6.plot()


plt.show()


```
