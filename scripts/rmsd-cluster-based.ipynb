```python
# Data
df=pd.read_csv("/home/ahr/Desktop/spike_MD/clusters_cutoff.csv")

#rangos del eje x
xlim_1 = 0.15
xlim_2 = 0.34
#rangos del eje y
ylim = 40
space=ylim/2
# multiple line plots
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(111)
ax1.set_title("Gradient of cut-off value and cluster size", fontsize=38)   
ax1.set_xlabel('cutoff', fontsize=18) 
ax1.set_ylabel('# of cluster', fontsize=18) 
plt.ylim(0,ylim)
plt.xlim(xlim_1,xlim_2)
plt.plot( 'cutoff', 'Delta 21J (1)', data=df, marker='o', markerfacecolor='black', markersize=6, color='red', linewidth=3)
#plt.plot( 'cutoff', 'Delta 21J (2)', data=df, marker='o', markerfacecolor='black', markersize=6, color='blue', linewidth=3)
plt.plot( 'cutoff', 'Delta 21J (3)', data=df, marker='o', markerfacecolor='black', markersize=6, color='green', linewidth=3)
plt.plot( 'cutoff', 'omicron21K', data=df, marker='o', markerfacecolor='black', markersize=6, color='orange', linewidth=3)
plt.plot( 'cutoff', 'omicron21L', data=df, marker='o', markerfacecolor='black', markersize=6, color='cyan', linewidth=3)
plt.plot( 'cutoff', 'original', data=df, marker='o', markerfacecolor='black', markersize=6, color='gray', linewidth=3)
plt.yticks(range(0, ylim+1, 1))
plt.xticks(np.arange(xlim_1,xlim_2, step=0.01)) 


#lineas horizontales
plt.axhline(10)

# show legend
plt.legend()

# show graph
plt.show()

##print
#print("En el caso de la variante delta -inter-, el n√∫mero de clusters es considerablemente menor (901) debido a que la trayectoria tiene menos frames, ")
#print("de hecho estos son el mismo n√∫mero de clusters encontrados para el cutoff mas bajo evaluado, -901 para 0.05-")
#print("mientras que el resto de variantes, tienen 3001 (el n√∫mero de frames de la trayectoria) al cutoff 0.05")
```

```python
## Crear listas vacias del eje x y y para depositar datos en ellas despues.
x, y = [], []
## Abrir el archivo (variable) con el cual se trabaja
## depositar en listas x y y los datos extraidos del archivo de trabajo
with open("/home/ahr/Desktop/spike_MD/" + str(Folder) + "/xcluster_files/cluster_" + str(Var) + "_rbd_" + str(F) + "_" +str(Cutoff) + ".xvg") as f:
    for line in f:
        cols = line.split()

        if len(cols) == 2:
            x.append(float(cols[0]))
            y.append(float(cols[1]))
## calcular el porcentaje de estructuras de los valores del eje y (estructuras)
total_structures = sum(y)
for i in range(len(y)):
    y[i] *= 100 / total_structures
## Gr√†ficar los datos (linea)
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(111)
ax1.set_title("Cluster size", fontsize=38)     
ax1.set_xlabel('¬∑# de Cluster', fontsize=18) 
ax1.set_ylabel('porcentaje de estructuras por cluster (%)', fontsize=18) 
ax1.plot(x,y, label='the data', color='red', linewidth=2.5)
leg = ax1.legend()
plt.xticks(range(1,2+int(max(x)), 1))
plt.yticks(range(1,5+int(max(y)), 2))
plt.ylim()
plt.show()  

print("Variante y cadena: " + str(Var))
print("El numero de cluster para el cutoff " + str(Cutoff) + " es: " + str(max(x)))
print(" ")
print("El porcentaje de estructuras en el... ")
print("cluster 1 es: " + str(round(y[0], 2)) + " %")
print("cluster 2 es: " + str(round(y[1], 2)) + " %")
print("cluster 3 es: " + str(round(y[2], 2)) + " %")
print("cluster 4 es: " + str(round(y[3], 2)) + " %")
print("cluster 5 es: " + str(round(y[4], 2)) + " %")
print(" ")

primeros_tres_clusters = round(y[0] + y[1] + y[2])
print("El porcentaje de estructuras en los primeros tres clusters es: " +str(primeros_tres_clusters) + " %" )

```

```python
##dataset del rmsd
x,y = np.loadtxt("/home/ahr/Desktop/spike_MD/" + str(Folder) + "/rmsd_md_" + str(Var) + "_3rbds_" + str(ns) + "ns_noh2o_fit.xvg",comments=["@", "#"],unpack=True)
##convertir a pandas dframe
df = pd.DataFrame (x, columns = ['time'])
df["rmsd"] = y
#df.head()
#df.tail()
```

```python
#dataset de cluster
x1, y1 = [], []
## Abrir el archivo (variable) con el cual se trabaja
## depositar en listas x y y los datos extraidos del archivo de trabajo
with open("/home/ahr/Desktop/spike_MD/" + str(Folder) + "/xcluster_files/cluster_" + str(Var) + "_rbd_" + str(F2) + "_" +str(Cutoff) + ".xvg") as f:
    for line in f:
        cols = line.split()

        if len(cols) == 2:
            x1.append(float(cols[0]))
            y1.append(float(cols[1]))
#a√±adir al dframe previo como una nueva columna
df["cluster"] = y1
#df.head()
```

```python
##Extraer la informaci√≤n de cada cluster
ymax = int(1+max(y1))
cluster_rms = {}
for num in range(1, ymax):
    cluster_rms[num] = df.loc[df['cluster'] == num, ['time', 'rmsd']]
```

```python
fig = plt.figure(figsize=(18, 8))
ax1 = fig.add_subplot(111)
for j in cluster_rms:
    ax1.scatter(cluster_rms[j]['time'], cluster_rms[j]['rmsd'], s=30, marker="o", label='Cluster ' 
                +str(j))
plt.legend(bbox_to_anchor=(1, 1.2), ncol=4,fontsize=12)
ax1.set_title("RMSD", fontsize=38) 
ax1.set_ylabel('nm', fontsize=18)
ax1.set_xlabel('time (ns)', fontsize=18)
plt.ylim()
plt.xlim()
plt.show()
```

```python
##Extraer la informaci√≤n de cada cluster
ymax = int(1+max(y1))
cluster_cl = {}
for num in range(1, ymax):
    cluster_cl[num] = df.loc[df['cluster'] == num, ['time', 'cluster']]
 
fig = plt.figure(figsize=(18, 10))
ax1 = fig.add_subplot(111)
for j in cluster_cl:
    ax1.scatter(cluster_cl[j]['time'], cluster_cl[j]['cluster'], s=30, marker="o", label='Cluster ' 
                +str(j))
plt.legend(bbox_to_anchor=(1, 1.2), ncol=4,fontsize=12)
ax1.set_title("Clusters", fontsize=38) 
ax1.set_ylabel('nm', fontsize=18)
ax1.set_xlabel('time (ns)', fontsize=18)
plt.ylim()
plt.yticks(range(1,5+int(max(y1)), 1))
plt.xlim()
plt.show()
```
