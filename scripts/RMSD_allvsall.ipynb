```python
# libraries
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Create a datasets
#centroide 1 | rmsd
df1 = pd.DataFrame({'Original':[0, 1.66, 2.25, 1.27,1.62],
                   'Delta1':[1.66, 0, 2.02, 1.79,1.59],
                   'Delta3':[2.25, 2.02, 0, 2.38,1.95],
                   'Omicron21K':[1.27, 1.79, 2.38, 0,1.75],
                   'Omicron21L':[1.62, 1.59, 1.95,1.75,0]
                   },
                  index=['Original', 'Delta1', 'Delta3', "Omicron21K", "Omicron21L"])

#centroide 1 | tm-score
df2 = pd.DataFrame({'Original':[1, 0.91944, 0.87165, 0.95305,0.92676],
                   'Delta1':[0.92723, 1, 0.9066, 0.91989,0.93226],
                   'Delta3':[0.87879, 0.9066, 1, 0.86818,0.90517],
                   'Omicron21K':[0.95305, 0.91224, 0.86127, 1,0.91891],
                   'Omicron21L':[0.91900,0.93226,0.90517,0.9113,1]
                   },
                  index=['Original', 'Delta1', 'Delta3', "Omicron21K", "Omicron21L"])
```

```python
# plot the heatmaps
fig, axes = plt.subplots(1,2, figsize=(15,6), constrained_layout=True)
fig.suptitle("RMSD (izq) y TM-score (der) entre las estructuras centroide del cluster 1 de variantes \n del RBD de la spike del SARS-CoV-2 \n", fontsize=22)
axes[0].set_title('RMSD')
axes[1].set_title('TM-score')

panel_a = sns.heatmap(df1, ax=axes[0],
        annot=True,
        annot_kws={"size": 15},
        vmin=0,
        vmax=2.38,
        cmap="coolwarm",
        center=1.19,
        linewidths=8,
        fmt = '',
        cbar_kws={"orientation": "horizontal"})

panel_b = sns.heatmap(df2, ax=axes[1],
        annot=True,
        annot_kws={"size": 15},
        vmin=0.8,
        vmax=1,
        cmap="coolwarm_r",
        center=0.96,
        linewidths=8,
        fmt = '',
        cbar_kws={"orientation": "horizontal"}
        )
        


panel_a.set_yticklabels(panel_a.get_yticklabels(), rotation=0)
panel_b.set_yticklabels(panel_b.get_yticklabels(), rotation=0)

```
